DIRECT METHOD FOR NORMAL EQUATION 



theta=(x.t,x)^-1.(x.t,y)


```python
import numpy as np

X = np.array([[2,3,5],[1,4,6],[7,8,9]], dtype=float)
y = np.array([4,8,12], dtype=float).reshape(-1,1)

# Normal equation
theta_closed = np.linalg.inv(X.T @ X) @ (X.T @ y)
# If X.T @ X may be singular, use pseudo-inverse:
theta_pinv = np.linalg.pinv(X) @ y
print("Theta (closed form):\n", theta_closed)
print("Theta (pseudo-inverse):\n", theta_pinv)
```
